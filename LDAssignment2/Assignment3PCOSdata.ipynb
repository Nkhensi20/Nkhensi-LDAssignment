```python
#Empirical rule (68-95-99.7)

mean = np.mean(dataset['BMI'])
std = np.std(dataset['BMI'])
def z_score(value, mean, std):
    return (value - mean) / std

```

```python
#random sampling

import random
values = []
for i in list(range(0,5)):
    value = random.choice(dataset.BMI)
    values.append(value)
print(values)
```

```python
def plot_hist(variable):
    plt.figure(figsize = (9,3))
    plt.hist(dataset[variable], bins = 50,color=color[0])
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title("{} distribution with hist".format(variable))
    plt.show()
    
```

```python
color = ['#B22222', '#FF7D40']

numericalVariables = [" Age (yrs)", "Weight (Kg)", "BMI", "Marraige Status (Yrs)"]
for n in numericalVariables:
    plot_hist(n)
```
